```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
img = cv2.imread('koala.JPG',0)
```

```
plt.imshow(img,cmap = 'gray')
print(img.shape)
```

```
img2 = cv2.imread('noisyimg.png',0)
plt.imshow(img2,cmap = 'gray')
print(img2.shape)
```

```
kernel = np.ones((5,5),np.float32)/25
```

```
dst = cv2.filter2D(img2,-1,kernel)
```

```
plt.imshow(dst,cmap='gray')
```

```
kernel = np.ones((3,3),np.float32)/9
dst = cv2.filter2D(img2,-1,kernel)
plt.imshow(dst,cmap='gray')
```

```
kernel = np.ones((7,7),np.float32)/49
dst = cv2.filter2D(img2,-1,kernel)
plt.imshow(dst,cmap='gray')
```

```
kernel = np.ones((9,9),np.float32)/81
dst = cv2.filter2D(img2,-1,kernel)
plt.imshow(dst,cmap='gray')
```

# New Section

```
blur = cv2.GaussianBlur(img2,(5,5),0)
plt.imshow(blur,cmap='gray')
```

```
median = cv2.medianBlur(img2,5)
plt.imshow(median,cmap='gray')
```

```
bi = cv2.bilateralFilter(img2,9,75,75)
plt.imshow(bi,cmap='gray')
```
